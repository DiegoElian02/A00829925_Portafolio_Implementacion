```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Valhalla23.csv")
df.head()
```

```python
plt.rcParams['figure.figsize'] = [10, 10]
plt.scatter(df['Celsius'], df['Valks'])
plt.show()
```

```python
X = df['Celsius']
Y = df['Valks']
```

```python
alpha = 0.001
thetas = [1, 1]
iteraciones = 10000

# Crear función lambda para la función de hipótesis
h = lambda thetas,x: thetas[0] + thetas[1]*x 

# Calcular el total de muestras a partir de los datos (n)
n = len(X)

for i in range(iteraciones):
    delta = []
    deltaX = []
    # Calcular delta para theta0 y para cada muestra
    for xi,yi in zip(X,Y):
        delta.append(h(thetas, xi)-yi)
    # Calcular delta para theta1 y para cada muestra
        deltaX.append((h(thetas,xi)-yi)*xi)
    # Calcular sumatorias y promedio

    dt0 = 1/n*sum(delta)
    dt1 = 1/n*sum(deltaX)
    # Actualizar theta0
    thetas[0] = thetas[0]- alpha*dt0
    # Actualizar theta1
    thetas[1] = thetas[1]- alpha*dt1
```

```python
thetas
```

```python
x = np.linspace(-20, 80, 100)

# Compute y values using the linear equation
y = thetas[1] * x + thetas[0]

# Create the plot
plt.figure(figsize=(10, 10))
plt.plot(x, y, label=f'y = {thetas[1]}x + {thetas[0]}', color='r')
plt.scatter(df['Celsius'], df['Valks'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of Linear Equation')
plt.legend()
plt.grid(True)
plt.show()
```
